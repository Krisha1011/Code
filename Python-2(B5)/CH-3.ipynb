```python
df = pd.DataFrame({
    'name':['A','B','C','D','E'],
    'Department':['CS','IT','EC','CS','IT']
    })
print(df)
print()
df_encode=pd.get_dummies(df,columns=['Department'])
print(df_encode)
```

```python
df = pd.DataFrame({
    'name':['A','B','C','D','E'],
    'Department':['CS','IT','EC','CS','IT']
    })
#print(df)
df_encode=pd.get_dummies(df,columns=['Department'],drop_first=True)
print(df_encode)
```

```python
import numpy as np

df = pd.DataFrame({
    'name':['A','B','C','D','E'],
    'Department':['CS','IT','EC','CS','IT'],
    'Score':[60,45,78,90,34]
    })

df['Grades']=np.where(df['Score']>60,'A','B')
print(df)
```

```python
df = pd.DataFrame({
    'name':['A','B','C','D','E'],
    'Department':['CS','IT','EC','CS','IT'],
    'Score':[60,45,78,90,34]
    })

df['Grades']=np.where(df['Score']>60,'A',np.where(df['Score']>40,'B','C'))
print(df)
```
