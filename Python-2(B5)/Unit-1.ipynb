```python
import pandas as pd
import numpy as np

data={"id":[1,2,2,3,4],"Name":["Amit","Priya",None,"Rahul",None],"Marks":[80,None,75,85,None],"City":["Ahmedabad","Rajkot",None,"Surat","Rajkot"]}
df=pd.DataFrame(data)
print(df)
print()
print(df.isnull())
print()
print(df.isnull().sum())
```

```python
df.fillna(130)
```

```python
df.fillna(df.mean(numeric_only=True))
```

```python
df.fillna(df.median(numeric_only=True))
```

```python
df.fillna(df.mode(numeric_only=True))
```

```python
df.fillna(df.mode().iloc[0]) # Fill NAN value alphabatically and if any value is reapeat then it fill with that.
```

```python
df.fillna(method="ffill") #Forward fill
```

```python
df.fillna(method="bfill") #Backward fill
```

```python
df.dropna()
```

```python
df.dropna(how='any')  
```

```python
df.dropna(how='all') # drop row which has all NAN Values
```

```python
df.dropna(how='any',axis=1) # drop column 
```

```python
df.dropna(subset=['Marks','City'])
```

```python
print(df.dropna(thresh=2)) # if non-null value is >=2 then it keeps that row
print()
print(df.dropna(thresh=3)) # if non-null value is >=3 then it keeps that row
print()
print(df.dropna(thresh=4)) # if non-null value is >=4 then it keeps that row
```

```python
data={"values":[0,40,35,45,32,39,42,100]}
df=pd.DataFrame(data)
print(df)
```

```python
Q1=df['values'].quantile(0.25)
print(Q1)
#Q2=df['values'].quantile(0.50)
#print(Q2)
Q3=df['values'].quantile(0.75)
print(Q3)

IQR=Q3-Q1 #Inter Qurtile Range
print(IQR) 

lower_bound=Q1-1.5*IQR
print(lower_bound)

upper_bound=Q3+1.5*IQR
print(upper_bound)

outliers=df[(df['values']<lower_bound)|(df['values']>upper_bound)]
print(outliers)
```

```python
clean_df=df[(df['values']>=lower_bound)&(df['values']<=upper_bound)]
print(clean_df)
```

```python
# concat()
import pandas as pd

df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
print(pd.concat([df1,df2]))
print()
print(pd.concat([df1,df2],axis=1))
```

```python
# join

df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'C':[7,8]})

print(pd.concat([df1,df2],join='outer',ignore_index=True))
```

```python
# merge()  need 1 column common

emp=pd.DataFrame({'emp_id':[1,2,3,4],'name':['A','B','C','D'],'dept_id':[10,20,10,30]})
dept=pd.DataFrame({'dept_id':[10,20,40],'name':['HR','IT','Finance']})

pd.merge(emp,dept,on='dept_id',how='inner')
```

```python
pd.merge(emp,dept,on='dept_id',how='outer')
```

```python
pd.merge(emp,dept,on='dept_id',how='left')
```

```python
pd.merge(emp,dept,on='dept_id',how='right')
```

```python
pd.merge(emp,dept,how='cross')
```

```python
# if in df no common colums found then use left_on,right_on

emp=pd.DataFrame({'emp_id':[1,2,3,4],'name':['A','B','C','D'],'dept_eid':[10,20,10,30]})
dept=pd.DataFrame({'dept_id':[10,20,40],'name':['HR','IT','Finance']})

pd.merge(emp,dept,left_on='dept_eid',right_on='dept_id',how='outer',indicator=True,sort=True,copy=True,suffixes=['_emp','_dept'])  
```

```python
emp=pd.DataFrame({'emp_id':[1,2,3,4],'name':['A','B','C','D'],'dept_id':[10,20,10,30]})
dept=pd.DataFrame({'dept_id':[10,20,40],'name':['HR','IT','Finance']})

pd.merge(emp,dept,on='dept_id',left_index=True)
```
