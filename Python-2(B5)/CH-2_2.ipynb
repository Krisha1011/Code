```python
import plotly.express as px
import plotly.io as pio
import pandas as pd
import matplotlib.pyplot as plt
pio.renderers.default='notebook'
```

```python
df = pd.read_csv("social_media_engagement.csv")
```

```python
df['platform'].unique
```

```python
fig = px.bar(df,x='post_type',y='likes',width=600,height=400,hover_data=['post_day','sentiment_score'],color='platform',color_discrete_map={'Facebook':'skyBlue','Instagram':'Pink','Twitter':'Black'})
fig.show()
```

```python
fig = px.bar(df,x='post_type',y='likes',width=600,height=400,hover_data=['post_day','sentiment_score'],
             color='platform',color_discrete_sequence=['skyBlue','Pink','Black'],pattern_shape='sentiment_score',
             text_auto=True)
fig.show()
```

```python
fig = px.bar(df,y='post_type',x='likes',width=600,height=400,hover_data=['post_day','sentiment_score'],
             color='platform',color_discrete_sequence=['skyBlue','Pink','Black'],pattern_shape='sentiment_score',
             text_auto=True,orientation='h')
fig.show()
```
